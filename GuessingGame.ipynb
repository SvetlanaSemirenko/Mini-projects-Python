### Описание проекта: 
Программа генерирует случайное число в диапазоне от 1 до n и просит пользователя угадать это число. Если догадка пользователя больше случайного числа, то программа должна вывести сообщение 'Слишком много, попробуйте еще раз'. Если догадка меньше случайного числа, то программа должна вывести сообщение 'Слишком мало, попробуйте еще раз'. Если пользователь угадывает число, то программа должна поздравить его и вывести сообщение 'Вы угадали, поздравляем!. Когда число отгадано, программа должна показать количество попыток и предложить сыграть ещё раз.


### Составляющие проекта:

Целые числа (тип int);

Переменные;

Ввод / вывод данных (функции input() и print());

Условный оператор (if/elif/else);

Цикл while;

Бесконечный цикл;

Операторы break, continue;

Работа с модулем random для генерации случайных чисел.

```python
import random

print('Hello! Welcome to number guessing game!')
x = int(input('Try to guess a number from 1 to ...?  '))

def is_valid(n):
    return n.isdigit and int(n) in range(1, x + 1)

def try_to_game():
    count = 0
    num = random.randint(1, x)
    while True:
        print(f'Enter a number from 1 to {x}')
        n = input()
        if is_valid(n) == False:
            print(f"Let's try again. Enter a number from 1 to {x} ?")
            continue
        else:
            n = int(n)
            if n > num:
                print("Your number is higher than expected. Try again.")
                count += 1
            elif n < num:
                print("Your number is less than expected. Try again.")
                count += 1
            else:
                count += 1
                print(f'Congratulations! You guessed right in {count} tries!')
                break                

try_to_game()

while True:
    print('Do you want to play again? (yes/no)')
    answer = input()
    if answer.lower() == 'yes' or answer.lower() == 'y':
        try_to_game()
    elif answer.lower() == 'no' or answer.lower() == 'n':
        print('Thanks for playing. See you soon:)')
        break
    else:
        print("Sorry:(. I don't unerstand, we keep playing?")
    
```

```python
import random

print('Привет! Добро пожаловать в числовую угадайку:)')
x = int(input('Попробуй угадать число в диапазоне от 1 до?  '))

def is_valid(n):
    return n.isdigit() and int(n) in range(1, x+1)

def try_to_game():
    count = 0
    num = random.randint(1, x)
    while True:
        print(f'Введите число от 1 до {x}')
        n = input()
    
        if is_valid(n) == False:
            print(f'А может быть все-таки введем целое число от 1 до {x}?')
            continue
        else:
            n = int(n)
            if n > num:
                print('Слишком много, попробуйте еще раз')
                count += 1
            elif n < num:
                print('Слишком мало, попробуйте еще раз')
                count += 1
            else:
                count += 1
                print(f'Вы угадали за {count} попыток, поздравляем!')
                break

try_to_game()

while True:
    print('Хотите сыграть еще раз? (Да/Нет)')
    answer = input()
    if answer.lower() == 'да' or answer.lower() == 'д':
        try_to_game()
    elif answer.lower() == 'нет' or answer.lower() == 'н':
            print('Спасибо, что играли в числовую угадайку. Еще увидимся...')
            break
    else:
        print('Простите, я Вас не понял, так вы хотите поиграть?')
```
